##**Córdoba**

The city that is going to be studied is Cordoba (Córdoba),located in the southern Spanish region of Andalusia and with a population close to 330,000 people. (Ayto. Córdoba, 2019). 

**History**

Cordoba has a long history, having existed at least since the eight century BC. In the year 711 AD. it became the capital of what would become the Al-Andalus empire, growing to a population close to one million inhabitants. Many of the historic places that have been bringing tourists to the city until now have their origin in that historic period. After the decadence of the empire and its collapse the city was conquered by the northern Kingdom of Castilla in 1236 (Turismo de Córdoba, s.f.).

It is remarkable to know that, almost 800 years after that, the economy of the city has been running, in part, thanks to the people that come to visit the historic places of that period. The _Mezquita-catedral de Córdoba_ alone was visited by almost two million people in 2019 (Europa Press, 2019).

**Gastronomics**

Another important economic motor of the city is Gastronomics. Some of the most famous foods of Spain have its origin in the region. Which is why it is also an influential factor in tourism (Luque, A., 2020).

**The Spring festivals**

The moment in which the quarantine was declared makes it possible that it will be coincident with some of the most important festivities of the city, which are concentrated in April and May. Those festivities are responsible for bringing thousands of people to Córdoba, and some of the most important of those events have already been canceled or delayed (Ávalos, R. 2020).

It is accepted that the tourism is a fundamental axis for the economy of the city. That tourism has been notably affected by the current pandemic, which is expected to cause loses running to millions of euros for the local economy (Cordópolis, 2020). How that situation will shape the local economy in the next few years is yet to be seen.

#**Methodology** <a name="methodology"></a>

This project is directed towards sampling the economic activity of the city of Cordoba taking advantage of the kind of data that Foursquare collects.
In the first part, needed libraries and prepared data are going to be called. That data is what has been defined above as the Table of Districts, neighbourhoods and Coordinates of Cordoba.

Once it is done, and to take advantage of the division of the city into districts, dataframes with the total of neighbourhoods and dataframes with the neighbourhoods of each districts, the number of venues businesses sampled will be measured and compared, to look for areas with a higher concentration than the average of measured areas. This is expected to give a portrait of the economic activity of the city that is related to tourism.

After it is done, a process of clustering using the k-means algorithm to analyse the similarity and dissimilarity of the neighbourhoods of Cordoba. The number of clusters will be determined using the Elbow Method. The resulting clusters will be analysed and compared.

Once the analysis process is finished, the results will be discussed and a conclusion, if possible, will be portrayed.

###5. Obtaining information about each neighbourhood venues

Using Foursquare API, samples of the venues of the city are extracted using the centers of the neighbourhoods as central points, with a radius of 500 meters and a limit of 100 venues per neighbourhood. 

```python
CLIENT_ID = 'CLIENT ID'
CLIENT_SECRET = 'CLIENT SECRET'
VERSION = '20202020' # Foursquare API version
radius=500
LIMIT=100
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
            
        # create the API request URL for Cordoba
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results0 = requests.get(url).json()["response"]['groups'][0]['items']
        print(results0)
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results0])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Barrio', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue business', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
cordoba_venues = getNearbyVenues(names=areascordoba['Barrio'],
                                   latitudes=areascordoba['Latitude'],
                                   longitudes=areascordoba['Longitude']
                                  )

def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venuescentro_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
            
        # create the API request URL for Cordoba CENTRO
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results1 = requests.get(url).json()["response"]['groups'][0]['items']
        print(results1)
        # return only relevant information for each nearby venue
        venuescentro_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results1])

    nearby_venuescentro = pd.DataFrame([item for venuecentro_list in venuescentro_list for item in venuecentro_list])
    nearby_venuescentro.columns = ['Barrio', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue business', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venuescentro)
cordobacentro_venues = getNearbyVenues(names=areascentro['Barrio'],
                                   latitudes=areascentro['Latitude'],
                                   longitudes=areascentro['Longitude']
                                  )

def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venueslevante_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
            
        # create the API request URL for Cordoba LEVANTE
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results2 = requests.get(url).json()["response"]['groups'][0]['items']
        print(results2)
        # return only relevant information for each nearby venue
        venueslevante_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results2])

    nearby_venueslevante = pd.DataFrame([item for venuelevante_list in venueslevante_list for item in venuelevante_list])
    nearby_venueslevante.columns = ['Barrio', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue business', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venueslevante)
cordobalevante_venues = getNearbyVenues(names=areaslevante['Barrio'],
                                   latitudes=areaslevante['Latitude'],
                                   longitudes=areaslevante['Longitude']
                                  )

def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venuesnoroeste_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
            
        # create the API request URL for Cordoba NOROESTE
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results3 = requests.get(url).json()["response"]['groups'][0]['items']
        print(results3)
        # return only relevant information for each nearby venue
        venuesnoroeste_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results3])

    nearby_venuesnoroeste = pd.DataFrame([item for venuenoroeste_list in venuesnoroeste_list for item in venuenoroeste_list])
    nearby_venuesnoroeste.columns = ['Barrio', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue business', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venuesnoroeste)
cordobanoroeste_venues = getNearbyVenues(names=areasnoroeste['Barrio'],
                                   latitudes=areasnoroeste['Latitude'],
                                   longitudes=areasnoroeste['Longitude']
                                  )

def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venuesnortesierra_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
            
        # create the API request URL for Cordoba NORTESIERRA
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results4 = requests.get(url).json()["response"]['groups'][0]['items']
        print(results4)
        # return only relevant information for each nearby venue
        venuesnortesierra_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results4])

    nearby_venuesnortesierra = pd.DataFrame([item for venuenortesierra_list in venuesnortesierra_list for item in venuenortesierra_list])
    nearby_venuesnortesierra.columns = ['Barrio', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue business', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venuesnortesierra)
cordobanortesierra_venues = getNearbyVenues(names=areasnortesierra['Barrio'],
                                   latitudes=areasnortesierra['Latitude'],
                                   longitudes=areasnortesierra['Longitude']
                                  )

def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venuespnorte_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
            
        # create the API request URL for Cordoba PONIENTE - NORTE
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results5 = requests.get(url).json()["response"]['groups'][0]['items']
        print(results5)
        # return only relevant information for each nearby venue
        venuespnorte_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results5])

    nearby_venuespnorte = pd.DataFrame([item for venuepnorte_list in venuespnorte_list for item in venuepnorte_list])
    nearby_venuespnorte.columns = ['Barrio', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue business', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venuespnorte)
cordobapnorte_venues = getNearbyVenues(names=areaspnorte['Barrio'],
                                   latitudes=areaspnorte['Latitude'],
                                   longitudes=areaspnorte['Longitude']
                                  )


def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venuespsur_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
            
        # create the API request URL for Cordoba PONIENTE - SUR
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results6 = requests.get(url).json()["response"]['groups'][0]['items']
        print(results6)
        # return only relevant information for each nearby venue
        venuespsur_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results6])

    nearby_venuespsur = pd.DataFrame([item for venuepsur_list in venuespsur_list for item in venuepsur_list])
    nearby_venuespsur.columns = ['Barrio', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue business', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venuespsur)
cordobapsur_venues = getNearbyVenues(names=areaspsur['Barrio'],
                                   latitudes=areaspsur['Latitude'],
                                   longitudes=areaspsur['Longitude']
                                  )
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venuessur_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
            
        # create the API request URL for Cordoba SUR
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results7 = requests.get(url).json()["response"]['groups'][0]['items']
        print(results7)
        # return only relevant information for each nearby venue
        venuessur_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results7])

    nearby_venuessur = pd.DataFrame([item for venuesur_list in venuessur_list for item in venuesur_list])
    nearby_venuessur.columns = ['Barrio', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue business', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venuessur)
cordobasur_venues = getNearbyVenues(names=areassur['Barrio'],
                                   latitudes=areassur['Latitude'],
                                   longitudes=areassur['Longitude']
                                  )

def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venuessureste_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
            
        # create the API request URL for Cordoba SURESTE
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results8 = requests.get(url).json()["response"]['groups'][0]['items']
        print(results8)
        # return only relevant information for each nearby venue
        venuessureste_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results8])

    nearby_venuessureste = pd.DataFrame([item for venuesureste_list in venuessureste_list for item in venuesureste_list])
    nearby_venuessureste.columns = ['Barrio', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue business', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venuessureste)
cordobasureste_venues = getNearbyVenues(names=areassureste['Barrio'],
                                   latitudes=areassureste['Latitude'],
                                   longitudes=areassureste['Longitude']
                                  )
```

###11. Clustering of the neighbourhoods

In this step, Machine Learning tool K-means will be used for clustering the neighbourhoods from their venues characteristics. That classification can be compared with the classification of Districts and can also be used for a better understanding of the economics of the city.

Before doing the ML clustering, information on the most common venue business category for each neighbourhood is going to be generated.

```python
cordoba_dummies = pd.get_dummies(cordoba_venues[['Venue Category']], prefix="", prefix_sep="") 

# add neighborhood column back to dataframe
cordoba_dummies['Barrio'] = cordoba_venues['Barrio'] 

# move neighborhood column to the first column
fixed_columns = [cordoba_dummies.columns[-1]] + list(cordoba_dummies.columns[:-1])
cordoba_dummies = cordoba_dummies[fixed_columns]
```

```python
cordoba_dummies.shape
```

```python
cordoba_grouped = cordoba_dummies.groupby('Barrio').mean().reset_index()
cordoba_dummies
```

```python
# Group rows by neighborhood and take the mean as frequency of occurrence of each cateogry.
cordoba_grouped = cordoba_dummies.groupby('Barrio').mean().reset_index()
cordoba_grouped
```

In the next table the ten most common venue business category of each neighbourhood can be seen.

```python
num_top_venues = 10

# Function to sort venues in descending order
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Barrio']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Barrio'] = cordoba_grouped['Barrio']

for ind in np.arange(cordoba_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(cordoba_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted
```

It might be interesting to know the frequency of the most common categories in every neighbourhood.

```python
num_top_venues = 5

for hood in cordoba_grouped['Barrio']:
    print("----"+hood+"----")
    temp = cordoba_grouped[cordoba_grouped['Barrio'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

From the frequency information, it can be seen that _El Naranjo_ has very particular venue categories, compared to the other neighbourhoods. That might have an influence in the K-means clustering.

---



**K-Means clustering of neighbourhoods**

In order to decide a number of clusters for the clustering, the _Elbow method_ is going to be applied. Knowing that the number of neighbourhoods is high, a wide range of _K_ is going to be taken into account.

```python
cordoba_grouped_clustering = cordoba_grouped.drop('Barrio',1)
```

```python
distortions = []
K = range(1,40)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(cordoba_grouped_clustering)
    distortions.append(kmeanModel.inertia_)
```

```python
plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

From the elbow method, it can be seen that there is not an extreme change of tendency in any step. There is a slight change between 5 and 10. Knowing that the number of official districts is 8 and that it might be interesting knowing what neighbourhoods are different enough to fit their own category, k=9 is chosen.

```python
kclusters=9


# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(cordoba_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[:] 
```

```python

# add clustering labels
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

cordoba_merged = areascordoba

cordoba_merged = cordoba_merged.join(neighborhoods_venues_sorted.set_index('Barrio'), on='Barrio')

cordoba_merged.head() # check the last columns!
cordoba_merged = cordoba_merged.dropna()
```

Once the clustering is done, a map is created with the information of the distribution of the neighbourhoods in clusters.

```python
# create map
map_clusters = folium.Map(location=[latitude_cordoba, longitude_cordoba], zoom_start=13)

folium.TileLayer('cartodbpositron').add_to(map_clusters)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(cordoba_merged['Latitude'], cordoba_merged['Longitude'], cordoba_merged['Barrio'], cordoba_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-3],
        fill=True,
        fill_color='Grey',
        fill_opacity=0.9).add_to(map_clusters)
       
map_clusters
```
